- Almost everything in python inherits from nn.Module (base class for all neural network modules)
- Our models should subclass this class - nn.Module
- We can stack these modules on top of each other and can make progressively more complex neural networks

## **nn.Parameter**
- Is a kind of tensor which is considered a module parameter
- It is a tensor in itself that has a special property when used with nn.Module
- when assigned as module attributes they are automatically added to the list of its parameters
- In short using this means creating a parameter
- `requires_grad=True` is used when we need to keep track of gradients (keep track of gradients of weights and bias parameters and update them using GD and Backprop)

## Forward method to define the computation in the model
- `def forward(self, x: torch.Tensor) -> torch.Tensor:`
- It takes in a parameter `x` (input data/features) - it needs to be of type torch.tensor
- we created **known** parameters weights and bias, which started of as random numbers, then run it through this forward calculation
- It will adjust these parameters (weight and bias), which started off as random numbers, to resemble as close as possible to known paramaters
- It does so using gradient descent and Backprop

image.png

```python
# Creating a linear regression model class

class LinearRegressionModel(nn.Module):
  # Creating constructor
  def __init(self):
    super().__init__()
    self.weights = nn.Parameter(torch.randn(1, # <- start with random weights (this will get adjusted as the model learns)
                                            dtype = torch.float), # <- PyTorch loves float32 by default
                                            requires_grad = True) # by default is also true
    self.bias = nn.Parameter(torch.randn(1,
                                         dtype = torch.float), # <- PyTorch loves float32 by default
                                          requires_grad = True)

    # Forward method to define the computation in the model
    def forward(self, x: torch.Tensor) -> torch.Tensor:
      return self.weights * x + self.bias
```

## Making Data using the Linear Regression Formula

- A `Parameter` is something that a model learns
- For the variables in the above equation, setting:
  - `b` as `weight`
  - `a` as `bias`

### Creating known parameters

```python
weight = 0.7
bias = 0.3
```

### Creating data

```python
# Capital letters denote matrix or tensor
# lower case represents a vector

X = torch.arange(start=0, end=1, step=0.02)
X
```

```python
# adding an extra dimension in X
X = X.unsqueeze(dim=1)
X
```

```python
y = weight * X + bias
y
```

## Creating a Train Test Split

```python
split = int(0.8 * len(X))
split
```

```python
X_train, y_train = X[:split], y[: split] # upto the split
len(X_train), len(y_train)
```

```python
X_test, y_test = X[split:], y[split:]
len(X_test), len(y_test)
```

- The model we create is going to try and learn the relationship between `X_train` & `y_train` and then we will evaluate what it learns on `X_test` and `y_test`.

# Lesson 18 - Creating a Range of Tensors and Tensor-like

- `torch.arange()` create a tensor containing a sequence of values evenly spaced within a specified range.

```python
torch.arange(0,10)
```

```python
torch.arange(start=2,end=5)
```

```python
torch.arange(start=5, end=50, step =7)
```

## Creating Tensors like

- Replicate the shape of the tensor, but dont want to explicitly tell what shape should be

```python
zero_to_10 = torch.arange(0,11)
zero_to_10
```

```python
ten_zeros = torch.zeros_like(zero_to_10)
ten_zeros
```

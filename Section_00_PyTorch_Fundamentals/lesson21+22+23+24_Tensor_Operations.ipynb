# Lesson 22+23 - Matrix Multiplication (dot product)

- In Matrix Multiplication - every row multiplies with every column
- `torch.mm` is short for `torch.matmul`, we can use `@` also

## Vector dot product

```python
tensor_a = torch.tensor([5,7,20])
tensor_b = torch.tensor([3,5,8])
```

```python
print(tensor_a, "*", tensor_b)
print(f"Equals: {tensor_a * tensor_b}")
```

```python
# 2nd method
torch.matmul(tensor_a, tensor_b) # faster method
```

## Matrix dot product

```python
torch.matmul(torch.rand(3,4), torch.rand(4,3)) # The inner dimensions must match
# The resulting matrix has shape of outer dimensions
```

# Lesson 24 - Transposing a Tensor

- A Matrix can be transposed in order to avoid dimension errors in multiplication.

```python
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])
```

```python
torch.mm(tensor_A, tensor_B)
```

```python
tensor_A.shape, tensor_B.shape
```

```python
# Transpose
tensor_B.T
```

```python
torch.mm(tensor_A, tensor_B.T)
```

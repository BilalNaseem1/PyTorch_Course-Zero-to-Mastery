# Lesson 25 - Tensor Aggregation (min, max, mean, sum etc.)

```python
x = torch.arange(0, 100, step=10)
x
```

```python
# finding min
torch.min(x), x.min()
```

```python
# max
torch.max(x), x.max()
```

```python
# mean
torch.mean(x)
```

- for mean data type should be floating point of tensor

```python
# changing data type of x
x = x.type(torch.float32)
x.dtype
```

```python
torch.mean(x)
```

```python
# sum
torch.sum(x), x.sum()
```

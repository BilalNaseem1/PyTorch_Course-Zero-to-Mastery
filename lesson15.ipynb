# Tensors

- Tensors are multi-dimensional arrays that represent and store data
- Serving as the fundamental building blocks for neural network computations.
- PyTorch tensors are created using `torch.tensor()`

## Types of Tensors
1. Scalar
2. Vector
3. Matrix
4. Tensors

- **ndim:** Number of dimensions - For example, an image could have three dimensions: width, height, and color channels.
  - no. of pair of square brackets
- **shape:** tells you how big your data is in each of these directions. If you have an image with a shape of 128x128x3, it means the image is 128 pixels wide, 128 pixels tall, and has 3 color channels

### Scalar Tensors

```python
scalar = torch.tensor(7)
scalar
```

```python
scalar.ndim
```

```python
scalar.shape
```

means 0x0

### Vector Tensors

```python
vector = torch.tensor([7, 7])
vector
```

```python
vector.ndim
```

```python
vector.shape
```

means 1x2

### Matrix Tensors

```python
matrix = torch.tensor([[7, 8],
                      [9, 10]])
matrix
```

```python
matrix.ndim
```

```python
matrix.shape
```

means 2x2

### Tensors

```python
tensor = torch.tensor([[[1,2,3],
                        [3, 6, 9],
                        [4, 4, 5]]])
tensor
```

```python
tensor.ndim
```

```python
tensor.shape
```

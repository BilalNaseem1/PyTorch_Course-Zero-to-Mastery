# Lesson 16 - Initializing Random Tensors

- Random tensors are often used for initializing neural network weights, which helps break symmetries and introduce diversity in model training.
- They prevent the network from starting with the same weights each time, allowing it to learn more effectively and generalize to different data patterns.
- `Start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers...`

### 3x4 Random Matrix

```python
random_tensor_3x4 = torch.rand(3,4)
random_tensor_3x4
```

```python
random_tensor_3x4.ndim
```

```python
random_tensor_3x4.shape
```

### 1x3x4 Random Tensor

```python
random_tensor_1x3x4 = torch.rand(1,3,4)
random_tensor_1x3x4
```
